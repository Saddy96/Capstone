***Class Imbalance - Using Combination of Oversampling and undersampling***

```python
# Handling Class Imbalance

#Required Dependencies for handling class imbalance
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
# define oversampling strategy
over = RandomOverSampler(sampling_strategy=0.2, random_state= 42)

# fit and apply the transform
X1, y1 = over.fit_resample(X, y)

# define undersampling strategy
under = RandomUnderSampler(sampling_strategy=0.98, random_state= 42)
# fit and apply the transform
# Assignging the balanced data back to x and y
X_comb, y_comb = under.fit_resample(X1,y1)
```

```python
y.count()
```

```python
sns.countplot(y_comb, palette= 'magma');
```

```python
y_comb.value_counts(normalize=True)
```

```python
from sklearn.model_selection import train_test_split

# 30% of the data will be used as test and remaining 70% for training the model
X_train, X_test, y_train, y_test = train_test_split(X_comb, y_comb, test_size=0.3, random_state=42)
```

***Feature Selection - USING Random Forest Classifier***
